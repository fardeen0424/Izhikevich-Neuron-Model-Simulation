```python
import numpy as np
import matplotlib.pyplot as plt

def izhikevich_neuron(I, a=0.02, b=0.2, c=-65, d=6, tau=0.25):
    v = -65
    u = b * v
    spikes = []
    for t in np.arange(0, 1000, tau):
        v += tau * (0.04 * v**2 + 5 * v + 140 - u + I)
        u += tau * a * (b * v - u)
        if v >= 30:
            spikes.append(t)
            v = c
            u += d
    return spikes

I_values = np.arange(0, 41, 1)
spike_rates = []
for I in I_values:
    spikes = izhikevich_neuron(I)
    spike_rate = len([t for t in spikes if t >= 200]) / 8
    spike_rates.append(spike_rate)

plt.plot(I_values, spike_rates)
plt.xlabel('Input Current (I)')
plt.ylabel('Spike Rate (Hz)')
plt.title('Regular Spiking Izhikevich Neuron')
plt.show()
```
